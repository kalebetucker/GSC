# Stock Price Retrieval

Goal of this script is to pull historical stock prices for input securities. Default @ SP500. 

```python
# Import Packages
import os
import numpy as np
import pandas as pd
import yfinance as yf
from datetime import date
```

```python
# Return SP500 as list
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
table = table[0]
ticker_list = set(table['Symbol'].tolist())
```

```python
# Download stock prices for listed securities
today = str(date.today())
portfolio = yf.download(ticker_list, start="2018-01-01", end=today);
```

```python
# Set Prices into dataframe
prices = portfolio['Adj Close']
```

## Security Charting

Plot an input securities price movement on line graph w/ standard deviations

```python
lookup = input("Do you have a specific security you would like to review? ")
lookupprices = prices[lookup]
MA = prices[lookup].rolling(50).mean()
lookupprices.plot(figsize=(15,10));
MA.plot()
```
